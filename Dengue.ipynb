Question 4: Custom Functions and the apply() Method
1. Write a custom function that performs a specific operation (e.g., categorize values
based on thresholds).
2. Use the apply() method to apply this function:
• Apply the function to rows (axis=1).
• Apply the function to columns (axis=0).
3. Demonstrate the difference in results when changing the axis parameter.

```python
def dengue(x):
  if x < 1:
    print("Not Infected")
  elif x == 1:
    print("Infected but not severe")
  elif x == 2:
    print("Infected and need care and attention")
  elif x == 3:
    print("Infected and severe")
```

```python
check = df["Test Report"].apply(lambda x:dengue(x))
```
