4. Apply Func?on
• Write a custom func<on that:
o Categorizes passengers as Child if their age is less than 18, Adult if their age is
between 18 and 60, and Senior if their age is greater than 60.
• Apply this func<on to the Age column and create a new column called AgeCategory

```python
def Age(x):
  if x < 18:
    print("Child")
  elif x > 18 and x < 60:
    print("Adult")
  elif x > 60:
    print("Senior")

df["AgeCategory"] = (df["Age"].apply(Age)) # Not storing data
```
